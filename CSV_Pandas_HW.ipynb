Extra Credit

```python
top_5_players_pts = acc_df[['Player', 'PTS']].sort_values(by='PTS', ascending=False).head(5)
top_5_players_pts.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.barplot(x='Player', y='PTS', data=top_5_players_pts)
plt.title('Points for Top 5 Point Scorers')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="FG%", y="WS", data=acc_df)
plt.title('Player Win Shares by Field Goal Percentage')
plt.show()
```

```python
r = acc_df['FG%'].corr(acc_df['WS'])
print(r)
```
