# Exercise 3

3.1

```python
aggregation_pipeline = [
    {"$unwind": "$genres"},
    {"$group": {"_id": "$genres", "avg_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"avg_rating": -1}},
    {"$limit": 5}
]

aggregated_data = collection.aggregate(aggregation_pipeline)
for data in aggregated_data:
    pprint.pprint(data)
```

3.2

```python
aggregation_pipeline2 = [
    {"$group": {"_id": "$directors", "avg_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"avg_rating": -1}},
    {"$limit": 5}
]

dir_w_best_imdb_data = collection.aggregate(aggregation_pipeline2)
for data in dir_w_best_imdb_data:
    pprint.pprint(data)
```

3.3

```python
aggregation_pipeline_3 = [
{"$group": {"_id": "$year", "total_movies": {"$sum": 1}}},
{"$sort": {"_id": 1}}
]

num_movies_each_year = collection.aggregate(aggregation_pipeline_3)

for data in num_movies_each_year:
    pprint.pprint(data)
```
