```python
import json
import pandas as pd
import json
import requests
```

```python
stock=input().strip()

if(stock == ""):
  print("Please enter a stock.")
else:
  apikey=""

  url1 = "https://yfapi.net/v6/finance/quote"

  querystring = {"symbols":stock}

  headers = {
     'x-api-key': apikey
     }

  response = requests.request("GET", url1, headers=headers, params=querystring)

  url2 = "https://yfapi.net/v11/finance/quoteSummary/"+stock+"?lang=en&region=US&modules=financialData"

  querystring = {"symbols":stock}

  headers = {
     'x-api-key': apikey
     }

  response2 = requests.request("GET", url2, headers=headers, params=querystring)

  stock_json = response.json()
  stock_json2 = response2.json()
  if(len(stock_json['quoteResponse']['result']) == 0):
    print("Please enter a valid stock symbol")
  else:
    try:
      print(stock_json['quoteResponse']['result'][0]["symbol"] + "        " + stock_json['quoteResponse']['result'][0]["longName"] + "        Price: $" + str(stock_json['quoteResponse']['result'][0]["regularMarketPrice"])+ "          Target Mean Price: $" + str(stock_json2['quoteSummary']['result'][0]["financialData"]["targetMeanPrice"]["raw"])  +"           Fifty Two Week Low: $" + str(stock_json['quoteResponse']['result'][0]["fiftyTwoWeekLow"]))
    except:
      print("There is an issue with the stock entered. Please try again.")
      exit(1)
    ticker_name = stock_json['quoteResponse']['result'][0]["symbol"]
    long_name = stock_json['quoteResponse']['result'][0]["longName"]
    price = stock_json['quoteResponse']['result'][0]["regularMarketPrice"]
    target_mean_price = stock_json2['quoteSummary']['result'][0]["financialData"]["targetMeanPrice"]["raw"]
    fifty_two_week_low = stock_json['quoteResponse']['result'][0]["fiftyTwoWeekLow"]

    url3 = "https://yfapi.net/v1/finance/trending/US"

    region = "US"

    headers = {
      'x-api-key': apikey
       }

    response3 = requests.request("GET", url3, headers=headers, params=region)
    stock_json3 = response3.json()
    print("Five Current Trending Stocks in the US:  " + stock_json3['finance']['result'][0]["quotes"][0]["symbol"] +  "    " + stock_json3['finance']['result'][0]["quotes"][1]["symbol"] + "    " +  stock_json3['finance']['result'][0]["quotes"][2]["symbol"] +   "    " + stock_json3['finance']['result'][0]["quotes"][3]["symbol"]  +   "    " + stock_json3['finance']['result'][0]["quotes"][4]["symbol"])

    #source for df to csv:  https://www.geeksforgeeks.org/saving-a-pandas-dataframe-as-a-csv/
    stock_data = [[ticker_name, long_name, price, target_mean_price, fifty_two_week_low]]
    stock_df = pd.DataFrame(stock_data, columns = ["Ticker Name", "Long Name", "Current Market Price", "Target Mean Price", "Fifty Two Week Low"])
    display(stock_df)
    stock_df.to_csv('stock_info.csv')


```
